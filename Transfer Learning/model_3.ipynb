```python
import os
import shutil
from torchvision.datasets import Flowers102
```

```python
train_dataset = Flowers102(root='data', split='train', download=True)
val_dataset = Flowers102(root='data', split='val', download=True)
test_dataset = Flowers102(root='data', split='test', download=True)
```

```python
base_dir = 'flowers_yolo'

os.makedirs(base_dir, exist_ok=True)
os.makedirs(os.path.join(base_dir, 'train'), exist_ok =True)
os.makedirs(os.path.join(base_dir, 'val'), exist_ok =True)
os.makedirs(os.path.join(base_dir, 'test'), exist_ok=True)
```

Copying images to class folders

```python
def convert_dataset(dataset, split):
    split_dir = os.path.join(base_dir, split)

    for img_path, label in zip(dataset._image_files, dataset._labels):

        label = int(label)

        class_dir = os.path.join(split_dir, str(label))
        os.makedirs(class_dir, exist_ok=True)

        filename = os.path.basename(img_path)
        dest_path = os.path.join(class_dir, filename)

        shutil.copy(img_path, dest_path)
```

```python
convert_dataset(train_dataset, 'train')
convert_dataset(val_dataset, 'val')
convert_dataset(test_dataset, 'test')
```
